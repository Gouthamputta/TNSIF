```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'Students':['G','S','J','D'],
    'Marks':[22,32,42,18]
})

sns.barplot(x='Students', y='Marks', data=df)
plt.title("Student Marks")
plt.show()
```

```python
sns.scatterplot(x='Students',y ='Marks', data=df)
plt.title("Student Marks")
plt.show()
```

```python
sns.lineplot(x='Students',y='Marks',data = df)
```
