```python
import numpy as np
import pandas as pd
s=pd.Series([1,2,3,4,5,6])
print(s)
print(s.describe())
print(s.apply(lambda s:s*2))
print(s.map(lambda s:s**2))
```

```python
s2=pd.Series([1,2,3,4,5,6])
print(s2.iloc[3])
print(s2.iloc[-2])
print(s.loc[2])
print(s.loc[1:2])
greater_num = s2[s2<=5]
print(greater_num)
```

```python
print(s2.sum())
print(s2.cumsum())
print(s2.aggregate(['sum','std','mean']))
```
