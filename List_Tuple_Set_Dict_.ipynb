**LIST**

```python
list1=["Apple","Banana","Apple","orange"]
list1.append("grape")
print(list1)
list1.insert(2,"berry")
print(list1)
list1.remove("Banana")
print(list1)
list1.sort()
print(list1)
list1.reverse()
print(list1)
list1.index("orange")
list1.count("Apple")
```

Tuple

```python
t1=(1,2,3,4,2)
print(t1)
print(len(t1))
t=t1.count(2)
print(t)
t1.index(4)
```

Set

```python
setA={'arjun','meera','vijay','ananya','rahul'}
setB={'vijay','ananya','rahul','kiran','deepak'}

print(setA)
print(setB)

setA.add('task')
print(setA)

setA.remove('arjun')
print(setA)

union_set=setA.union(setB)
print(union_set)

intersection_set=setA.intersection(setB)
print(intersection_set)

difference_set=setA.difference(setB)
print(difference_set)

setA.clear()
print(setA)
```


Dictionary

```python
student={
    "name":"Bharath",
    "age":20,
    "course":"AIML"
}

print(student)
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))

student.pop('name')
print(student)

student.update({'name':'hello'})
print(student)

student.update({'age':23})
print(student)
```

Strings

```python
s='python programming'
print(s)
print(len(s))
print(s.upper())
print(s.lower())
print(s.capitalize())
print(s.title())
print(s.replace('python','java'))
print(s.split())
print(s.find('programming'))
print(s.count('o'))
print(s.startswith("Python"))
print(s.endswith("programming"))
```
