**Perceptron**

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

w = np.zeros(2)
b = 0
lr = 0.1

for _ in range(20):
    for i in range(len(X)):
        pred = 1 if np.dot(X[i],w)+b >= 0 else 0
        w += lr*(y[i]-pred)*X[i]
        b += lr*(y[i]-pred)

plt.scatter(X[:,0],X[:,1],c=y,s=100)
x = np.linspace(-0.5,1.5,100)
plt.plot(x,(-w[0]*x-b)/w[1])
plt.show()
```

#Activation Functions

#1. Sigmoid Activation Function



```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))  #where e = 2.718

study_hours = 5

weight = 2
bias = -8

z = (weight * study_hours) + bias

prediction = sigmoid(z)

print("Z Value :", z)
print("Sigmoid Output :", prediction)
```

#2. ReLU (Rectified Linear Unit)

```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

weight = 2
bias = -8

hours = [1,2,3,4,5,6,7,8]

print("Hours\tZ\tReLU")

for h in hours:
    z = weight*h + bias
    output = relu(z)

    print(f"{h}\t{z}\t{output}")
```

#3. Tanh (Hyperbolic Tangent)

```python
import numpy as np

def tanh(x):
    return np.tanh(x)

z_values = [-6, -2, 0, 2, 6]

outputs = tanh(z_values)

print("Neuron Outputs :", z_values)
print("After Tanh     :", outputs)
```

#4. Leaky ReLU


```python
import numpy as np

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

scores = [-10, -5, -2, 0, 2, 5, 10]

print("Score\tLeaky ReLU Output")

for score in scores:
    print(score, "\t", leaky_relu(score))
```

#5. Softmax

```python
import numpy as np

# Raw scores (logits)
scores = np.array([4.5, 2.0, 0.5])

# Softmax function
def softmax(x):
    exp_values = np.exp(x)
    return exp_values / np.sum(exp_values)

probabilities = softmax(scores)

print("Softmax Probabilities:")
print(probabilities)

classes = ["Dog", "Cat", "Bird"]

predicted = classes[np.argmax(probabilities)]

print("\nPredicted Class :", predicted)
```
