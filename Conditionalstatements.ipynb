IF Statement

If else Statement

```python
num=6
if(num%2==0):
  print("Even number")
else :
  print("odd number")
```

Else If Statements

```python
marks=90
if(marks<=100 and marks>=80):
  print("A grade")
elif (marks<80 and marks>=50):
  print("B grade")
elif(marks<50 and marks>=40) :
  print("C grade")
elif(marks<40 and marks>=0) :
  print("fail")
else :
  print("Enter valid input")
```

Nested If Statements

```python
age=20
citizen=True
if(citizen) :
  if(age>=18) :
    print("Eligible")
  else:
    print("not eligible")
```

```python
#to check the number is prime or not
num = int(input("Enter number"))
count=0
for i in range(1,num+1):
  if(num%i==0):
    count+=1
if(count==2):
  print("its a prime number")
else:
  print("not a prime")
```

program on factorial of a number

```python
digit=6
fact=1
for i in range(1,digit+1):
    fact=fact*i
print(fact)
```

program on palindrome or not

```python
num=123
temp=num
reverse=0
while num>0:
    reminder=num%10
    reverse=reverse*10+reminder
    num=num//10
print(reverse)
if temp==reverse:
    print("palindrome")
else:
    print("Not a palindrome")
```

Program to check Armstrong number o not

```python
num=543
total=0
temp=num
l=len(str(num))
for i in range(l):
    rem=num%10
    total=total+(rem**l)
    num=num//10
print(total)
if temp==total:
    print("Armstrong")
else:
    print("Not an Armstrong")
```

Program on pattern printing
